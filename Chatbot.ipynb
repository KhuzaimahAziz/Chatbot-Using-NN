```python
data={'intents': [{'tag': 'greeting',
   'patterns': ['Hi there',
    'How are you',
    'Is anyone there?',
    'Hey',
    'Hola',
    'Hello',
    'Good day'],
   'responses': ['Hello',
    'Good to see you again',
    'Hi there, how can I help?'],
   'context': ['']},
  {'tag': 'goodbye',
   'patterns': ['Bye',
    'See you later',
    'Goodbye',
    'Nice chatting to you, bye',
    'Till next time'],
   'responses': ['See you!', 'Have a nice day', 'Bye! Come back again soon.']},
  {'tag': 'thanks',
   'patterns': ['Thanks',
    'Thank you',
    "That's helpful",
    'Awesome, thanks',
    'Thanks for helping me'],
   'responses': ['My pleasure', "You're Welcome"]},
  {'tag': 'intention',
   'patterns': ['What Kind of help you can give me?',
    'Can you Guide me?',
    'I need help in Selecting Items for shopping',
    'Can you recommend me Somthing?'],
   'responses': ["Sure. I'm your AI Chatbot that can help you recommend Apparel. Can you Specify What sort of things you like"]},
  {'tag': 'help_query1',
   'patterns': ['Im a guy so I want somthing for Gym.',
    'I want you to recommend Gym clothings for men'],
   'responses': ['Sure. For men I recommend, Nike t-shirt and shorts. They are fashionable as well as good for Gym.']},
  {'tag': 'help_query2',
   'patterns': ["Would you recommend wearing Jeans to Gym?",'Should I wear Jeans to Gym?', 'Can jeans be worn in Gym for workout?'],
   'responses': ['No. I would not recommend wearing jeans to gym as it makes it harder for you to make movments while wearing it.']
   },
  {'tag': 'help_query3',
   'patterns': ['Im a women so I want somthing for Gym.','I want you to recommend Gym clothings for women'],
   'responses': ['Sure. For women I recommend, Tanktops or Sports attire from Nike and leggings. They are good for Gym for women.']
}]}
```

```python
with open('intents.json') as file:
    data = json.load(file)
```

```python
data
```

```python
training_sentences = []
training_labels = []
labels = []
responses = []


for intent in data['intents']:
    for pattern in intent['patterns']:
        training_sentences.append(pattern)
        training_labels.append(intent['tag'])
    responses.append(intent['responses'])
    
    if intent['tag'] not in labels:
        labels.append(intent['tag'])
```

```python
num_labels=len(labels)
```

```python
labels
```

```python
training_sentences
```

```python
encoder = LabelEncoder()
encoder.fit(training_labels)
training_labels = encoder.transform(training_labels)
```

```python
embedding_dim = 16
max_len = 20
oov_token = "<OOV>"

tokenizer = Tokenizer(num_words=1000+1, oov_token=oov_token)
tokenizer.fit_on_texts(training_sentences)
word_index = tokenizer.word_index
sequences = tokenizer.texts_to_sequences(training_sentences)
padded_sequences = pad_sequences(sequences, truncating='post', maxlen=max_len)
```

```python
word_index
```

```python
model = Sequential()
model.add(Embedding(1000, embedding_dim, input_length=max_len))
model.add(GlobalAveragePooling1D())
model.add(Dense(16, activation='relu'))
model.add(Dense(16, activation='relu'))
model.add(Dense(num_labels, activation='softmax'))

model.compile(loss='sparse_categorical_crossentropy', 
              optimizer='adam', metrics=['accuracy'])

model.summary()
```

```python

epochs = 500
history = model.fit(padded_sequences, np.array(training_labels), epochs=epochs)
```

```python
model.save("chat_model")
```

```python
import pickle

# to save the fitted tokenizer
with open('tokenizer.pickle', 'wb') as handle:
    pickle.dump(tokenizer, handle, protocol=pickle.HIGHEST_PROTOCOL)
    
# to save the fitted label encoder
with open('label_encoder.pickle', 'wb') as ecn_file:
    pickle.dump(encoder, ecn_file, protocol=pickle.HIGHEST_PROTOCOL)
```

```python
def chat():
    # load trained model
    model = keras.models.load_model('chat_model')

    # load tokenizer object
    with open('tokenizer.pickle', 'rb') as handle:
        tokenizer = pickle.load(handle)

    # load label encoder object
    with open('label_encoder.pickle', 'rb') as enc:
        lbl_encoder = pickle.load(enc)

    # parameters
    max_len = 20
    
    while True:
        print("User: ", end="")
        inp = input()
        if inp.lower() == "quit":
            break

        result = model.predict(keras.preprocessing.sequence.pad_sequences(tokenizer.texts_to_sequences([inp]),
                                             truncating='post', maxlen=max_len))
        tag = lbl_encoder.inverse_transform([np.argmax(result)])

        for i in data['intents']:
            if i['tag'] == tag:
                print("ChatBot:", np.random.choice(i['responses']))

        # print(Fore.GREEN + "ChatBot:" + Style.RESET_ALL,random.choice(responses))

print("Start messaging with the bot (type quit to stop)!")
chat()
```
